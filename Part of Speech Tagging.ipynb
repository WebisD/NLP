# POS Tagging



*   **(N) Noun**- Daniel, London, table, dog, teacher, pen, city, happiness, hope
*   **(V) Verb**- go, speak, run, eat, play, live, walk, have, like, are, is
*   **(ADJ) Adjective**- big, happy, green, young, fun, crazy, three
*   **(ADV) Adverb**- slowly, quietly, very, always, never, too, well, tomorrow
*   **(P) Preposition**- at, on, in, from, with, near, between, about, under
*   **(CON) Conjunction**- and, or, but, because, so, yet, unless, since, if
*   **(PRO) Pronoun**- I, you, we, they, he, she, it, me, us, them, him, her, this
*   **(INT) Interjection**- Ouch! Wow! Great! Help! Oh! Hey! Hi!
*   **(NUM) Numeral**- twenty-four, fourth, 1991, 14:24



```
s1 = "Apple is looking at buying U.K. startup for $1 billion"
```

```
import spacy
nlp = spacy.load(name='en_core_web_sm')
```

```
doc = nlp(s1)
```

```
for token in doc:
  print(token.text.ljust(10), token.pos_.ljust(10), token.tag_.ljust(10), spacy.explain(token.tag_))
```

```
for key, val in doc.count_by(spacy.attrs.POS).items():
  print(key, doc.vocab[key].text, val)
```

```
from spacy import displacy
```

```
displacy.render(docs=doc, style='dep', options={'distance':110}, jupyter=True)
```
