## 2) Data Cleaning

*   Remove Stopwords
*   Remove punctuations
*   Lemmatization



```
import string
```

```
punct = string.punctuation
```

```
punct
```

```
from spacy.lang.en.stop_words import STOP_WORDS
```

```
stopwords = list(STOP_WORDS) # list with all the stop words
```

```
print(stopwords)
```

```
import spacy
nlp = spacy.load('en_core_web_sm')
```

```
def text_data_cleaning(sentence):
  doc = nlp(sentence)

  tokens = [] # list of tokens
  for token in doc:
    if token.lemma_ != "-PRON-":  # Se não for pronome
      temp = token.lemma_.lower().strip()
    else: #  Se for, já está na forma orignal
      temp = token.lower_
    tokens.append(temp)

  cleaned_tokens = []
  for token in tokens:
    if token not in stopwords and token not in punct:
      cleaned_tokens.append(token)
  return cleaned_tokens
```

```
text_data_cleaning("Hello all of you, It´s a beautiful day outside there!")
```

## Vectorization Feature Engineering (TF-IDF)

-> term frequency - inverse document frequency

Classifica uma palavra baseado no quanto ela aparece no documento com a quantidae de documentos que ela aparece

*   Palavras raras terão pontuação mais alta
*   Palavras comuns terão pontuação mais baixa

```
from sklearn.svm import LinearSVC
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.pipeline import Pipeline
```

```
tfidf = TfidfVectorizer(tokenizer=text_data_cleaning)
# Tokenization será feito a partir da função 'text_data_cleaning' e depois fará o TF-IDF
```

```
classifier = LinearSVC()
```

## 3) Train the Model

```
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=0)  # 20% para teste
```

```
x_train.shape, y_test.shape
```

```
x_train.head()
```

### Fit the x_train and y_train

```
clf = Pipeline([('tfidf', tfidf), ('clf', classifier)])  #  clf possivel relacionamento com supor vector machine svm
# first do : Vectorization // second do : classification
```

```
clf.fit(x_train, y_train)
```

## 4) Predict the Test set results

```
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```
y_pred = clf.predict(x_test)
```

```
confusion_matrix(y_test, y_pred)

# acertou 201 errou 78
# errou 50 acertou 221
```

```
print(classification_report(y_test, y_pred))

# 77% de acuracia :P
```

```
accuracy_score(y_test, y_pred)

# 76% de acuracia
```
