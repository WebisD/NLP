# 1) Matplotlib Part 1

## 1) Functional Method

```
import matplotlib.pyplot as plt
import numpy as np
from numpy.random import randint
```

```
x = np.linspace(0,10,20) # 0 to 10, catch 20 numbers
```

```
x
```

```
y = randint(0,50,20) # 0 to 50, catch 20 random numbers
```

```
y
```

```
y = np.sort(y)
```

```
y
```

```
plt.plot(x, y, color='m', linestyle=':', marker='*', markersize=10, lw=2)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('X vs Y axis')
plt.show()  # Show the graph
```

### 1.1) Multiple plots on same canvas

```
plt.subplot(1,2,1) 
# Numbers of rows
# Number of columns 
# The plot position
plt.plot(x,y,color='r')
plt.subplot(1,2,2)
plt.plot(np.sqrt(y),x,color='c')
```

## 2) Object Oriented Method

```
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import randint
```

```
x = np.linspace(0,10,20)
y = randint(1,50,20)
y = np.sort(y)
```

```
x
```

```
y
```

```
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,1,1]) # left/bottom/width/height
axes.plot(x,y)
axes.set_xlabel('X axis')
axes.set_ylabel('Y axis')
axes.set_title('X vs Y axis')
```

### 2.1) 2 sets of figures to 1 canvas

```
fig = plt.figure()
ax1 = fig.add_axes([0.1,0.1,0.8,0.8])
ax2 = fig.add_axes([0.3,0.4,0.4,0.3])

ax1.plot(x,y,color='g')
ax1.set_xlabel('X axis')
ax1.set_ylabel('Y axis')
ax1.set_title('Plot 1')

ax2.plot(x,y,color='m')
ax2.set_xlabel('X axis')
ax2.set_ylabel('Y axis')
ax2.set_title('Plot 2')
```

# 2) Matplotlib Part 2

### 2.1) Subplots method

```
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import randint
```

```
x = np.linspace(0,10,20)
y = randint(1,50,20)
y = np.sort(y)
```

```
x
```

```
y
```

```
fig, axes = plt.subplots()
axes.plot(x,y)
```

```
fig, axes = plt.subplots(nrows=1, ncols=2)
```

```
fig, axes = plt.subplots(nrows=2, ncols=3)
plt.tight_layout() # No overlapping
```

```
axes
```

```
fig, axes = plt.subplots(nrows=1, ncols=2)
for current_ax in axes:
  current_ax.plot(x,y)
```

```
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y)
axes[1].plot(x,y**2)
axes[0].set_title('Plot 1')
axes[1].set_title('Plot 2')
```

### 2.2) Figure size, Aspect ration and DPI

```
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import randint
```

```
x = np.linspace(0,10,20)
y = randint(1,50,20)
y = np.sort(y)
```

```
#               Width and height
fig = plt.figure(figsize=(5,2), dpi=100)
# DPI = Dot per inch // Resolution
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(x,y)
```

### 2.3) Legends

```
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label='X vs X square')
ax.plot(x,x**3, label='X vs X cube')
ax.plot(x,x**4, label='X vs X quad')
ax.legend(loc=10) # loc = position
```

# 3) Matplotlib Part 3

```
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import randint
```

```
x = np.linspace(0,10,20)
y = randint(1,50,20)
y = np.sort(y)
```

```
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color='c', lw=3, ls='-.', alpha=0.9, marker='o', markersize=10, markerfacecolor='y')
```

```
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='r', lw=3)
ax.set_xlim([0,1])
ax.set_ylim([0,10])
```

# 4) Different Plots

```
import matplotlib.pyplot as plt
```

## 4.1) Scatter Plots

```
y_views=[534,690,258,402,724,689,352]
f_views=[123,342,700,305,406,648,325]
t_views=[202,209,176,415,824,389,550]
days=[1,2,3,4,5,6,7]
```

```
plt.scatter(days, y_views, label='Youtube Views', marker='o')
plt.scatter(days, f_views, label='Facebook Views', marker='o')
plt.scatter(days, t_views, label='Twitter Views', marker='o')
plt.xlabel('Days')
plt.ylabel('Views')
plt.title('Social Media Views')
plt.grid(color='r', ls='--')
plt.legend()
```

## 4.2) Bar Plot

```
plt.bar(days, y_views, label='Youtube Views')
plt.bar(days, f_views, label='Facebook Views')
#plt.bar(days, t_views, label='Twitter Views')
plt.xlabel('Days')
plt.ylabel('Views')
plt.title('Social Media Views')
plt.legend()
```

## 4.3) Histogram

```
points=[22,55,62,45,21,22,99,34,42,4,102,110,27,48,99,84]
bins=[0,20,40,60,80,100,120]
```

```
plt.hist(points, bins)
plt.xlabel('Bins')
plt.ylabel('Frequency')
plt.title('Bins vs Frequency')
plt.show()
```

## 4.4) Pie chart

```
labels_1=['Facebook','Instagram','Youtube','linkedin']
views=[300,350,400,450]
explode_1=[0,0,0,0.2]
```

```
plt.pie(views, labels=labels_1, autopct='%1.1f%%', explode=explode_1, shadow=True)
plt.show()
```
