## Layers


<pre>

### 1) Input Layer

In a image problem, it holds the raw input of a image. With width, height and depth

### 2) Convolution Layer

Mathematical operation that takes the:
* Image matrix
* Filter (kernel)

then it's multiplied the image matriz with the filter genrating the output

### 3) ReLU Layer

**ReLU** = Rectified Linear Unit

It's an activation function with the folowuing charcteristcs:

* Equation: A(x) = max(0,x)  // Sempre valores >= 0
* Non-linear function
* Less expensive than the others activation functions because only a few neurons are activated

### 4 Pooling Layer

Usefull to:

* decrease the computacional power required to process data
* Extract dominant features wich are rotational and positional invariant

### 5) Fully connected layer

The final that returns the output layer

All nodes are conected to all of the nextlayer
