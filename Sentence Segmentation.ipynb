# Sentence Segmentation

```
s1 = "This is a sentence. This is a second sentence. This is last sentence."
s2 = "This is a sentence; This is a second sentence; This is last sentence."
```

```
import spacy
nlp = spacy.load(name='en_core_web_sm')
```

```
doc1 = nlp(s1)
```

```
for sent in doc1.sents: #  Sents = Sentences
  print(sent.text)
```

```
s3 = "This is a sentence. This is a second U.K. sentence. This is last sentence."
```

```
doc3 = nlp(s3)
for sent in doc3.sents:
  print(sent.text)
```

```
doc2 = nlp(s2)
for sent in doc2.sents:
  print(sent.text)
```

```
def set_custom_boundaries(doc):
  for token in doc[:-1]:
    if token.text == ';':
      print(token.i)
      doc[token.i+1].is_sent_start = True
  return doc
```

```
nlp.pipe_names
# Steps: 
# 1 - Tagging
# 2 - Parsing
# 3 - Name entity recognizer
```

```
nlp.add_pipe(set_custom_boundaries, before='parser')
```

```
nlp.pipe_names
```

```
newDoc2 = nlp(s2)
for sent in newDoc2.sents:
  print(sent.text)
```
