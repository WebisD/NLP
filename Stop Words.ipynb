# Stop Words

Ou tambem conhecidas como palavras vazias, são palavras irrelevantes ou que não adicionam muito contexto à frase. Normalmente são as palavras mais comuns em uma lingua e podem ser retiradas sem que a frase perca o sentido.

Ex: as, e, os, de, para, com, sem // a, an, the

```
import spacy
nlp = spacy.load('en_core_web_sm')
```

```
print(nlp.Defaults.stop_words)
```

```
len(nlp.Defaults.stop_words)
```

```
nlp.vocab['however'].is_stop
```

```
nlp.vocab['finance'].is_stop
```

## Add Stop Word

```
nlp.vocab['LOL'].is_stop
```

```
nlp.Defaults.stop_words.add('LOL')
```

```
nlp.vocab['LOL'].is_stop
```

```
nlp.vocab['LOL'].is_stop = True
```

```
nlp.vocab['LOL'].is_stop
```

## Remove Stop Word

```
nlp.vocab['no'].is_stop
```

```
nlp.vocab['no'].is_stop = False
```

```
len(nlp.Defaults.stop_words)
```

```
print(nlp.Defaults.stop_words)
```

```
nlp.Defaults.stop_words.remove('no')
```

```
nlp.vocab['no'].is_stop
```

```
nlp.vocab['no'].is_stop = False
```
