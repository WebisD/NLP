## Stemming and Lemmatization


### Stemming


> Mais simples

>Reduzir a palavra ao seu "tronco" ou "raiz", tirando o radical dela. Mas nem sempre a definição que aparece no dicionário. remove o sufixo e extrai o radical
>> Ex: Playing, Played, Play => Play.

>> Ex: Aprender => Aprend / Cursar => Curs

```
words = ['run','runner','running','ran','runs','easily','fairly']
```

```
import nltk
```

```
from nltk.stem.porter import PorterStemmer
from nltk.stem.snowball import SnowballStemmer
```

```
p_stemmer = PorterStemmer()
s_stemmer = SnowballStemmer(language='english') # A little better than porter
```

```
for word in words:
  print(word + " => " + p_stemmer.stem(word))
```

```
for word in words:
  print(word + " => " + s_stemmer.stem(word))
```

## Lemmatization

> Encontrar o Lema de uma palavra, ou seja, uma palavra base, aquela que encontramos no dicionário. Análise morfológica
>Pode manter o significado das palavras, como nomes.
>>Ex: Aprendeu, aprendendo -> Aprender

```
import spacy
nlp = spacy.load('en_core_web_sm')
```

```
doc1 = nlp("The striped bats are hanging on their feet for best")
```

```
for token in doc1:
  print(token.text, '\t', token.lemma_)
```

```
doc = nlp("play played playing do did done doing")
[token.lemma_ for token in doc]
```

```
s1 = "The striped bats are hanging on their feet for best"

for word in s1.split():
  print(word + " => " + p_stemmer.stem(word))
```
